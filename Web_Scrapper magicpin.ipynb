```python
import requests
from bs4 import BeautifulSoup as bs
import pandas as pd
```

```python
# link='https://www.amazon.in/Apple-iPhone-14-128GB-Midnight/product-reviews/B0BDHX8Z63/ref=cm_cr_dp_d_show_all_btm?ie=UTF8&reviewerType=all_reviews'
# link = 'https://www.amazon.in/Redmi-Jade-Black-6GB-128GB/product-reviews/B0C9JFWBH7/ref=cm_cr_dp_d_show_all_btm?ie=UTF8&reviewerType=all_reviews'
link = 'https://magicpin.in/New-Delhi/Rajinder-Nagar/Restaurant/M-Gopinath-South-Indian/store/3aa900/reviews/'
```

```python
headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 6.1) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/41.0.2228.0 Safari/537.36'
}

page = requests.get(link,headers=headers)
```

```python
page
```

```python
soup = bs(page.content,'html.parser')
```

```python
print(soup.prettify())
```

```python
names = soup.find_all('p',class_='name')
```

```python
names
```

```python
#Extracting customer names
cust_name = []

for i in range(len(names)):
    cust_name.append(names[i].text.strip())

cust_name
```

```python
len(cust_name)
```

```python
#Extracting Review title.
followers = soup.find_all('p',class_='followers')
```

```python
followers
```

```python
review_title = []
for i in range(0,len(followers)):
    review_title.append(followers[i].text.strip())
review_title
```

```python
review_title[:] = [titles.lstrip('\n') for titles in review_title]
review_title
```

```python
review_title[:] = [titles.rstrip('\n') for titles in review_title]
review_title
```

```python
len(review_title)
```

```python
rating = soup.find_all('p',class_='review')
rating
```

```python
rate = []
for i in range(0,len(rating)):
    rate.append(rating[i].text.strip())
rate
```

```python
len(rate)
```

```python
df = pd.DataFrame()
df['Customer Name']=cust_name[:10]
df['Followers']=review_title[:10]
df['Ratings']=rate[:10]
```

```python
df
```

```python
df.to_csv(r'./reviews.csv',index=True)
df
```
